```python
# imports
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import PIL as pil
import cv2
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"]=20,20
```

```python
# load images
cwd = os.getcwd()
# base_image_path = os.path.join(cwd, 'Images', 'Beispiel 1_Hintergrundbild.jpg')
test_image_path = os.path.join(cwd, 'Images', 'test_1.png')
# base_image_path = os.path.join(cwd, 'Images', '0000.jpg')
# test_image_path = os.path.join(cwd, 'Images', '0010.jpg')
```

```python
image = cv2.imread(test_image_path)
imgray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grads = np.gradient(imgray)
grad = np.abs(grads[0]) + np.abs(grads[1])
grad[grad < 0.5 * grad.max()] = 0.0
grad = np.interp(grad, (grad.min(), grad.max()), (0.0, 255.0)).astype(np.uint8)
# img = pil.Image.fromarray(grad)
# img.show()
```

```python
#grad = cv2.cvtColor(grad, cv2.COLOR_BGR2GRAY)
imgray = imgray ** 1.0
imgray = np.interp(imgray, (imgray.min(), imgray.max()), (0.0, 255.0)).astype(np.uint8)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
_, contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image, contours, -1, (0,255,0), 3)
img = pil.Image.fromarray(image)
img.show()
# cv2.imshow('Contours', image)
```
